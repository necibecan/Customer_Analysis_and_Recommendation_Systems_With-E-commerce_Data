```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('comp.filter.csv')
```

```python
df.dropna(inplace= True)
```

```python
df['user_score'] = df['event_type'].map({'view':1,'cart':10,'purchase':50})
```

```python
nokta = "."
df['Description'] = df['brand'] + nokta + df['category_code']
```

```python
df
```

```python
df.drop('Unnamed: 0',axis=1, inplace=True)
df.drop('Unnamed: 0.1',axis=1, inplace=True)
```

```python
#df.to_csv('recommender_df.csv')
```

```python
df["product_id"].nunique()
```

```python
df["product_id"].value_counts().head()
```

```python
comment_counts = pd.DataFrame(df["product_id"].value_counts())
rare_product = comment_counts[comment_counts["product_id"] <= 1000].index
```

```python
common_product = df[~df["product_id"].isin(rare_product)]
user_product_df = common_product.pivot_table(index=["user_id"], columns=["product_id"], values="user_score")
```
